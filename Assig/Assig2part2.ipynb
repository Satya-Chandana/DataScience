```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1.	Read the csv file as DataFrame.
df = pd.read_csv(
    'https://github.com/flaviovdf/evcomp2018/raw/master/data/nba_salaries.csv  ')
df
```

```python
#2) Find size 
df.size
```

```python
#3) Sort on player 
df.sort_values("PLAYER")

```

```python
#4) Sort on salary(increasing and decreasing) 
df.sort_values("SALARY")
df.sort_values("SALARY", ascending=False)
```

```python
#5) Relabel column salary as “SALARY: 
df.rename(columns={'salary': 'SALARY'}, inplace=True)
df
```

```python
#6) Display first 5 rows.
df.head(5)
```

```python
#7)Select 3rd, 10th and 22nd rows.
df.iloc[3]
df.iloc[10]
df.iloc[22]
```

```python
#8) Top 5 highest paid.
df.sort_values("SALARY", ascending=False).head(5)
```

```python
#9)List those earned more than 10m USD.
df[df.SALARY > 10]
```

```python
#10)Earnings of Stephen Curry
df.loc[df["PLAYER"] == "Stephen Curry"]["SALARY"].values[0]
```

```python
#11) All Point Guards(PG) salary more than 15m USD.
df.loc[(df["POSITION"] == "PG")&( df["SALARY"] > 15 )]["PLAYER"].values
```

```python
#12)Team “ Golden State Warriors”

df.loc[(df["TEAM"] == "Golden State Warriors")]["PLAYER"].values
```

```python
#13)How much money did each team pay for its players' salaries? (We have to group the rows by TEAM and then sum the salaries of the groups.
df.groupby("TEAM").sum()

```

```python
#14)How many NBA players were there in each of the five positions? (Classify by POSITION , and count – use grouping).
df.groupby("POSITION").count()
```

```python
#15)What was the average salary of the players at each of the five positions?(Group by POSITION and take the mean of the salaries)
df.groupby("POSITION").mean()
```
